```python
import pandas as pd

# 打开第一个文件（包含要删除的行）
with open("./cool_list.txt", "r") as file1:
    lines1 = file1.readlines()

new_lines = []

for line in lines1:
    parts = line.strip()  # 使用strip()方法删除末尾的换行符
    new_lines.append("/home/cp/Desktop/GSE146937/out_cool/" + parts)

sub_df1 = pd.DataFrame()
sub_df1[0] = new_lines

sub_df1.to_csv("./cool_path.txt", sep='\t', index=None, header=None, mode='a')
```
